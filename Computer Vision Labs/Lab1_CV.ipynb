<h2> read the image using skimage and opencv </H2>

```python
# read image
image_skim_rgb = io.imread(r'C:\Users\user\Downloads\adventure_time_tree_house_sunset_scenery_4k_wallpaper_uhdpaper_c.jpg')

image_cv_gb = cv2.imread(r'C:\Users\user\Downloads\adventure_time_tree_house_sunset_scenery_4k_wallpaper_uhdpaper_c.jpg')
```

<h2>convart the image to rgb</h2>

```python
image_cv_rgb = cv2.cvtColor(image_cv_gb, cv2.COLOR_BGR2RGB)

plt.imshow(image_cv_rgb)
```

# show the image after convert gray scale 


```python
photo_grayScale = color.rgb2gray(image_skim_rgb)

plt.title('the image in gray scale')
io.imshow(photo_grayScale)
```

```python
from skimage import filters # get gaussain filter  from skimage
```

```python
#show the image after applay gausian filter using skimage

gaus_ski = filters.gaussian(image_skim_rgb,channel_axis=-1)

#show the image after applay gausian filter using cv2

gaus_cv = cv2.GaussianBlur(image_cv_rgb,(3,3),0)
```

```python

fig = plt.figure()
fig.add_subplot(2,2,1)
plt.imshow(gaus_ski)
plt.title('after gaussion filter in skimage')

fig.add_subplot(2,2,2)
plt.imshow(gaus_cv)
plt.title('after gaussion filter in CV2')

```

```python
from skimage import feature


```

```python
canny_sk = feature.canny(photo_grayScale,sigma=0.9)
#io.imshow(canny_sk)
canny_cv = cv2.Canny(cv2.cvtColor(image_cv_rgb,cv2.COLOR_BGR2RGB),70,70 ,apertureSize=5,L2gradient=True)

fig = plt.figure()
fig.add_subplot(2,2,1)
plt.imshow(canny_sk)
plt.title('using skimage')

fig.add_subplot(2,2,2)
plt.imshow(canny_cv)
plt.title('using openCV')
```
